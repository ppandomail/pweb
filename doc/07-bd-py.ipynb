## BD SQLite

### Crear BD y abrir conexión

```python
import sqlite3
con = sqlite3.connect('mydatabase.db')
# Si no existe la crea
```

### Crear cursor para ejecutar consultas SQL

```python
cur = con.cursor()
```

### Crear tabla

```python
cur.execute('CREATE TABLE customers(id, name, address)')
```

```python
cur.execute('CREATE TABLE cities(cod, name)')
```

### Mostrar tablas

```python
res = cur.execute('SELECT * FROM sqlite_master')
res.fetchall()
```

### Insertar fila

```python
cur.execute("INSERT INTO customers VALUES (1, 'pablo', 'capilla del señor')")
con.commit()
```

### Insertar filas

```python
cur.execute("""
    INSERT INTO customers VALUES
        (2, 'mariano', 'palermo'),
        (3, 'ana', 'merlo')
""")
con.commit()
```

### Insertar filas con marcadores de posición

```python
data = [
    (4, 'luis', 'bariloche'),
    (5, 'maria', 'iguazú'),
    (6, 'alex', 'rosario'),
]
cur.executemany('INSERT INTO customers VALUES(?, ?, ?)', data)
con.commit()
```

### Seleccionar datos

```python
res = cur.execute('SELECT name FROM customers')
res.fetchall()
```

```python
for row in cur.execute('SELECT name, address FROM customers'):
    print(row)
```

### Crear conexión a la BD

```python
mydb = mysql.connector.connect(
  host='localhost',
  user='yourusername',
  password='yourpassword',
  database='mydatabase'
)
```

### Crear tabla

```python
mycursor = mydb.cursor()
mycursor.execute('CREATE TABLE customers (name VARCHAR(255), address VARCHAR(255))')
```

### Mostrar tablas

```python
mycursor.execute('SHOW TABLES')
for x in mycursor:
  print(x)
```

### Modificar tabla

```python
mycursor.execute('ALTER TABLE customers ADD COLUMN id INT AUTO_INCREMENT PRIMARY KEY')
```

### Insertar fila

```python
sql = 'INSERT INTO customers (name, address) VALUES (%s, %s)'
val = ('John', 'Highway 21')
mycursor.execute(sql, val)
mydb.commit()
print(mycursor.rowcount, 'record inserted.')
```

### Insertar filas

```python
val = [
  ('Peter', 'Lowstreet 4'),
  ('Amy', 'Apple st 652'),
  ('Hannah', 'Mountain 21'),
  ('Michael', 'Valley 345'),
  ('Sandy', 'Ocean blvd 2'),
  ('Betty', 'Green Grass 1'),
  ('Richard', 'Sky st 331'),
  ('Susan', 'One way 98'),
  ('Vicky', 'Yellow Garden 2'),
  ('Ben', 'Park Lane 38'),
  ('William', 'Central st 954'),
  ('Chuck', 'Main Road 989'),
  ('Viola', 'Sideway 1633')
]
mycursor.executemany(sql, val)
mydb.commit()
print(mycursor.rowcount, 'was inserted.')
```

### Seleccionar datos

```python
mycursor.execute('SELECT * FROM customers')
myresult = mycursor.fetchall()
for x in myresult:
  print(x)
```

### Seleccionar columnas

```python
mycursor.execute('SELECT name, address FROM customers')
myresult = mycursor.fetchall()
for x in myresult:
  print(x)
```

### Seleccionar primera fila del resultado

```python
mycursor.execute('SELECT * FROM customers')
myresult = mycursor.fetchone()
myresult
```

### Seleccionar con filtro

```python
sql = "SELECT * FROM customers WHERE address ='Park Lane 38'"
mycursor.execute(sql)
myresult = mycursor.fetchall()
for x in myresult:
  print(x)
```

### Seleccionar con comodín

```python
sql = "SELECT * FROM customers WHERE address LIKE '%way%'"
mycursor.execute(sql)
myresult = mycursor.fetchall()
for x in myresult:
  print(x)
```

### Ordenar resultado

```python
sql = 'SELECT * FROM customers ORDER BY name'
mycursor.execute(sql)
myresult = mycursor.fetchall()
for x in myresult:
  print(x)
```

### Ordenar en orden inverso

```python
sql = 'SELECT * FROM customers ORDER BY name DESC'
mycursor.execute(sql)
myresult = mycursor.fetchall()
for x in myresult:
  print(x)
```

### Eliminar filas

```python
sql = "DELETE FROM customers WHERE address = 'Mountain 21'"
mycursor.execute(sql)
mydb.commit()
print(mycursor.rowcount, 'record(s) deleted')
```

### Eliminar una tabla

```python
sql = 'DROP TABLE customers'
mycursor.execute(sql)
```

### Eliminar una tabla solo si existe

```python
sql = 'DROP TABLE IF EXISTS customers'
mycursor.execute(sql)
```

### Modificar filas

```python
sql = "UPDATE customers SET address = 'Canyon 123' WHERE address = 'Valley 345'"
mycursor.execute(sql)
mydb.commit()
print(mycursor.rowcount, 'record(s) affected')
```

### Limitar el resultado

```python
mycursor.execute('SELECT * FROM customers LIMIT ')
myresult = mycursor.fetchall()
for x in myresult:
  print(x)
```

### Limitar el resultado comenzando desde otra posición

```python
mycursor.execute('SELECT * FROM customers LIMIT 5 OFFSET 2')
myresult = mycursor.fetchall()
for x in myresult:
  print(x)
```

## MongoDB

* Bases de datos NoSQL
* Almacena los datos en documentos similares a JSON (BD flexibles y escalables)

1. Instalar [MongoDB] ( https://www.mongodb.com) o [Servicio Nube](https://www.mongodb.com/cloud/atlas) 
2. python -m venv venv
3. . .\venv\Scripts\activate
4. pip install pymongo

### Crear conexión a la BD

```python
import pymongo

myclient = pymongo.MongoClient('mongodb://localhost:27017/')
mydb = myclient['mydatabase']
# En MongoDB, una base de datos no se crea hasta que recibe contenido
```

### Mostrar BDs

```python
myclient.list_database_names()
```

### Crear Colección: en MongoDB es lo mismo que una tabla en las bases de datos SQL

```python
mycol = mydb['customers']
# En MongoDB, una colección no se crea hasta que recibe contenido
```

### Mostrar colecciones

```python
mydb.list_collection_names()
```

### Insertar un documento: en MongoDB es lo mismo que un registro en las bases de datos SQL

```python
mydict = { "name": "John", "address": "Highway 37" }
x = mycol.insert_one(mydict)
x.inserted_id
```

### Insertar muchos documentos

```python
mylist = [
  { "_id": 1, "name": "John", "address": "Highway 37"},
  { "_id": 2, "name": "Peter", "address": "Lowstreet 27"},
  { "_id": 3, "name": "Amy", "address": "Apple st 652"},
  { "_id": 4, "name": "Hannah", "address": "Mountain 21"},
  { "_id": 5, "name": "Michael", "address": "Valley 345"},
  { "_id": 6, "name": "Sandy", "address": "Ocean blvd 2"},
  { "_id": 7, "name": "Betty", "address": "Green Grass 1"},
  { "_id": 8, "name": "Richard", "address": "Sky st 331"},
  { "_id": 9, "name": "Susan", "address": "One way 98"},
  { "_id": 10, "name": "Vicky", "address": "Yellow Garden 2"},
  { "_id": 11, "name": "Ben", "address": "Park Lane 38"},
  { "_id": 12, "name": "William", "address": "Central st 954"},
  { "_id": 13, "name": "Chuck", "address": "Main Road 989"},
  { "_id": 14, "name": "Viola", "address": "Sideway 1633"}
]
x = mycol.insert_many(mylist)
x.inserted_ids
```

### Seleccionar primer documento del resultado

```python
x = mycol.find_one()
x
```

### Seleccionar todos los documentos

```python
for x in mycol.find():
  print(x)
```

### Devolver solo los nombres y direcciones, no los _ids

```python
for x in mycol.find({},{ "_id": 0, "name": 1, "address": 1 }):
  print(x)
# Se produce un error si se especifican los valores 0 y 1 en el mismo objeto 
# (excepto si uno de los campos es el campo _id)
```

### Filtrar el resultado

```python
myquery = { 'address': 'Park Lane 38' }
mydoc = mycol.find(myquery)
for x in mydoc:
  print(x)
```

### Filtrar con modificadores

```python
myquery = { 'address': { '$gt': 'S' } }   # $gt: mayor que
mydoc = mycol.find(myquery)
for x in mydoc:
  print(x)
```

### Filtrar con expresiones regulares

```python
myquery = { 'address': { '$regex': '^S' } }
mydoc = mycol.find(myquery)
for x in mydoc:
  print(x)
```

### Ordenar el resultado

```python
mydoc = mycol.find().sort('name')
for x in mydoc:
  print(x)
```

### Ordenar el resultado en orden inverso

```python
mydoc = mycol.find().sort('name', -1)
for x in mydoc:
  print(x)
```

### Eliminar documento

```python
myquery = { 'address': 'Mountain 21' }
mycol.delete_one(myquery)
#Si la consulta encuentra más de un documento, solo se elimina la primera aparición
```

### Eliminar muchos documentos

```python
myquery = { 'address': {'$regex': '^S'} }
x = mycol.delete_many(myquery)
print(x.deleted_count, ' documents deleted.')
```

### Eliminar todos los documentos de una colección

```python
x = mycol.delete_many({})
print(x.deleted_count, ' documents deleted.')
```

### Eliminar colección

```python
mycol.drop()
```

### Modificar un documento

```python
myquery = { 'address': 'Valley 345' }
newvalues = { "$set": { 'address': 'Canyon 123' } }
mycol.update_one(myquery, newvalues)
for x in mycol.find():
  print(x)
```

### Modificar muchos documentos

```python
myquery = { 'address': { '$regex': '^S' } }
newvalues = { '$set': { 'name': 'Minnie' } }
x = mycol.update_many(myquery, newvalues)
print(x.modified_count, 'documents updated.')
```

### Limitar el resultado

```python
myresult = mycol.find().limit(5)
for x in myresult:
  print(x)
```
