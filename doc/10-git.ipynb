# Guía de Git

* Git es el estándar de la industria para el control de versiones
* A continuación, una explicación detallada de los conceptos, herramientas y comandos que se usan a diario

## 1. ¿Qué es Git?

* **Git** es un software de control de versiones **local**.
* Su función es guardar por commits el código a medida que se avanza en el mismo (nuevos cambios).
* Si algo se rompe, se puede volver a cualquier punto (commit) del pasado.
* Es **distribuido**, lo que significa que cada desarrollador tiene una copia completa del historial en su propia máquina.

## 2. Repositorios

Aunque Git funciona en una computadora (local), se necesita un lugar en la nube para compartir el código y colaborar.

* **GitLab:** muy popular en entornos empresariales. Ofrece herramientas potentes de DevOps, CI/CD integrado y permite instalarlo en servidores propios (Self-hosted).
* **Bitbucket:** propiedad de Atlassian. Se integra a la perfección con **Jira** y **Confluence**. Es muy utilizado en equipos que ya usan el ecosistema de Atlassian para gestión de proyectos.
* **GitHub:** la plataforma más grande y usada del mundo, especialmente para proyectos de código abierto (Open Source).

## 3. Comandos de Configuración e Inicio

* **`git init`**: inicializa un nuevo repositorio de Git en la carpeta donde estás parado. Crea la carpeta oculta `.git`.
* **`git clone <url>`**: descarga un repositorio existente desde GitLab, Bitbucket o GitHub a tu máquina.

## 4. El Ciclo de Confirmación (Add & Commit)

Este es el flujo constante mientras programas:

* **`git add <archivo>`**: mueve los cambios al **Staging Area** (el área de preparación). Es como decir: "estos archivos están listos para la foto". Usá `git add .` para incluir todo lo modificado.
* **`git commit -m "mensaje"`**: saca la "foto" oficial. Guarda los cambios del Staging Area en el historial local con un mensaje explicativo.

## 5. Inspección y Estado

* **`git status`**: dice qué archivos están modificados, cuáles están listos para el commit y en qué rama se encuentra trabajando. Es el comando que más vas a usar para no perderte.
* **`git log`**: muestra el historial de commits. Se puede ver quién hizo qué, cuándo y el ID único (hash) de cada cambio.

## 6. Ramas y Navegación (Branch & Checkout)

* **`git branch <nombre>`**: crea una nueva rama. Las ramas permiten trabajar en una funcionalidad sin afectar el código principal.
* **`git checkout <nombre-rama>`**: te "mueve" de una rama a otra. También podés usar `git checkout <id-commit>` para volver temporalmente a una versión anterior del código.

## 7. Sincronización (Push & Pull)

* **`git push origin <rama>`**: sube tus commits locales al servidor remoto (GitLab/Bitbucket).
* **`git pull`**: descarga los cambios que tus compañeros subieron al servidor y los fusiona con tu código local.

## 8. Comandos de Emergencia y Utilidad

* **`git stash`**: ¿tenés que cambiar de rama pero tus cambios actuales están a medias? El "stash" guarda tus cambios en un lugar temporal y limpia el directorio para que puedas moverte sin hacer commit. Usá `git stash pop` para recuperarlos después.
* **`git revert <id-commit>`**: si se sube algo que rompe alguna funcionalidad o es algo que no se deseaba subir, el revert crea un nuevo commit que deshace exactamente lo que hizo el commit anterior. Es la forma más segura de corregir errores en repositorios compartidos.

## 9. Resumen de Flujo Típico

1.  `git pull`
2.  `git status`
3.  `git add .`
4.  `git commit -m "fix error modal login"`
5.  `git push`
