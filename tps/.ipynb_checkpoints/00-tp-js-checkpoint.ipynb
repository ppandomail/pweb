## Etapa 1

```javascript
const express = require('express')
const swaggerJsdoc = require('swagger-jsdoc')
const swaggerUi = require('swagger-ui-express')

const app = express()
app.use(express.json())

const productos = [
  {id:1, nombre:'python', precio:10.0},
  {id:2, nombre:'javascript', precio:12.0},
  {id:3, nombre:'java', precio:19.0}
];

let carrito = {}

const swaggerOptions = {
  definition: {
    openapi: '3.0.0',
    info: {
      title: 'API Carrito de Compras',
      version: '1.0.0',
      description: 'Documentación de la API Carrito de Compras'
    },
  },
  apis: ['./app.js'], // rutas donde están los comentarios JSDoc
}

const swaggerSpec = swaggerJsdoc(swaggerOptions)
app.use('/api-docs', swaggerUi.serve, swaggerUi.setup(swaggerSpec))

/**
 * @swagger
 * /productos:
 *   get:
 *     summary: Obtener lista de productos
 *     responses:
 *       200:
 *         description: Lista de productos disponibles
 */
app.get('/productos', (req, res) => {
  res.json(productos)
})

/**
 * @swagger
 * /carrito:
 *   get:
 *     summary: Obtener contenido del carrito
 *     responses:
 *       200:
 *         description: Carrito actual
 */
app.get('/carrito', (req, res) => {
  res.json(carrito)
})

/**
 * @swagger
 * /agregar/{id}:
 *   post:
 *     summary: Agregar producto al carrito
 *     parameters:
 *       - in: path
 *         name: id
 *         required: true
 *         schema:
 *           type: integer
 *         description: ID del producto
 *     responses:
 *       200:
 *         description: Carrito actualizado
 */
app.post('/agregar/:id', (req, res) => {
  const id = parseInt(req.params.id)
  const producto = productos.find(p => p.id === id)

  if (!producto) 
    return res.status(404).json({ error: 'Producto no encontrado' })
  
  carrito[id] = (carrito[id] || 0) + 1
  res.json(carrito)
})

/**
 * @swagger
 * /eliminar/{id}:
 *   delete:
 *     summary: Eliminar producto del carrito
 *     parameters:
 *       - in: path
 *         name: id
 *         required: true
 *         schema:
 *           type: integer
 *         description: ID del producto
 *     responses:
 *       200:
 *         description: Carrito actualizado
 */
app.delete('/eliminar/:id', (req, res) => {
  const id = parseInt(req.params.id)

  if (!carrito[id]) 
    return res.status(404).json({ error: 'Producto no está en el carrito' })

  if (carrito[id] > 1) 
    carrito[id] -= 1
  else 
    delete carrito[id]

  res.json(carrito);
});

/**
 * @swagger
 * /vaciar:
 *   delete:
 *     summary: Vaciar el carrito
 *     responses:
 *       200:
 *         description: Carrito vaciado
 */
app.delete('/vaciar', (req, res) => {
  carrito = {}
  res.json({ mensaje: 'Carrito vaciado', carrito })
})

/**
 * @swagger
 * /total:
 *   get:
 *     summary: Calcular total de la compra
 *     responses:
 *       200:
 *         description: Total acumulado
 */
app.get('/total', (req, res) => {
  let total = 0.0
  for (const [id, cantidad] of Object.entries(carrito)) {
    const producto = productos.find(p => p.id === parseInt(id))
    if (producto)
      total += producto.precio * cantidad
  }
  res.json({ total })
})

const PORT = 3000
app.listen(PORT, () => {
  console.log(`Servidor corriendo en http://localhost:${PORT}`)
})
```

[**swagger**: http://localhost:3000/api-docs/](http://localhost:3000/api-docs/)

```javascript
const URL_BASE = 'http://127.0.0.1:3000'

async function test_status_code_ok() {
  await fetch(`${URL_BASE}/vaciar`, { method: 'DELETE' })
  const res = await fetch(`${URL_BASE}/productos`)
  console.log('status_code_ok:', res.status === 200)
}

async function test_cantidad_productos() {
  await fetch(`${URL_BASE}/vaciar`, { method: 'DELETE' })
  const res = await fetch(`${URL_BASE}/productos`)
  const data = await res.json()
  console.log('cantidad_productos:', data.length === 3)
}

async function test_agregar_mismo_producto() {
  await fetch(`${URL_BASE}/vaciar`, { method: 'DELETE' })
  await fetch(`${URL_BASE}/agregar/1`, { method: 'POST' })
  await fetch(`${URL_BASE}/agregar/1`, { method: 'POST' })
  const res = await fetch(`${URL_BASE}/carrito`)
  const data = await res.json()
  console.log('agregar_mismo_producto:', Object.keys(data).length === 1)
}

async function test_agregar_distintos_productos() {
  await fetch(`${URL_BASE}/vaciar`, { method: 'DELETE' })
  await fetch(`${URL_BASE}/agregar/1`, { method: 'POST' })
  await fetch(`${URL_BASE}/agregar/2`, { method: 'POST' })
  const res = await fetch(`${URL_BASE}/carrito`)
  const data = await res.json()
  console.log('agregar_distintos_productos:', Object.keys(data).length === 2)
}

async function test_total() {
  await fetch(`${URL_BASE}/vaciar`, { method: 'DELETE' })
  await fetch(`${URL_BASE}/agregar/1`, { method: 'POST' })
  await fetch(`${URL_BASE}/agregar/2`, { method: 'POST' })
  const res = await fetch(`${URL_BASE}/total`)
  const data = await res.json()
  console.log('total:', data.total === 22)
}
```

```javascript
(async () => {
  await test_status_code_ok()
  await test_cantidad_productos()
  await test_agregar_mismo_producto()
  await test_agregar_distintos_productos()
  await test_total()
})()
```
