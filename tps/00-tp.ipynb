# TP: Desarrollo de Aplicación Web SPA
---

## Objetivo

Desarrollar una aplicación web Single Page Application (SPA) inspirada en la aplicación [Coffee Cart](https://coffee-cart.app/), en dos etapas progresivas:
1. Backend con APIs
2. Frontend, base de datos y pruebas E2E

## Etapa 1: Backend con APIs

* **Descripción**: Implementar un servidor backend que exponga una serie de APIs RESTful para gestionar un carrito de compras de ...
* **Requerimientos mínimos**:
    * Endpoints para:
        * Listar productos disponibles
        * Agregar productos al carrito
        * Eliminar productos del carrito
        * Calcular el total de la compra
    * Persistencia inicial en memoria (sin base de datos)
    * Documentación de las APIs (por ejemplo, con Swagger/OpenAPI)
    * Tests unitarios de los endpoints principales

## Etapa 2: Frontend + BD + Pruebas E2E

* **Descripción**: Construir el frontend SPA que consuma las APIs del backend y agregar persistencia con base de datos
* **Requerimientos mínimos**:
    * Frontend en React, Vue, Angular o HTML vainilla
    * Interfaz que permita:
        * Visualizar productos
        * Agregar/eliminar productos del carrito
        * Mostrar el total de la compra
    * Integración con una base de datos (ej. SQLite, PostgreSQL, MySQL, MongoDB)
    * Pruebas E2E (ej. con Selenim, Cypress o Playwright) que validen:
        * Flujo de compra completo
        * Persistencia de datos
        * Correcta interacción entre frontend y backend
    * Deploy en un entorno accesible (ej. Vercel, Netlify, Render, Railway)

### Entregables

1. **Etapa 1**: Repositorio con backend + documentación de APIs + tests unitarios
2. **Etapa 2**: Repositorio con frontend + integración con base de datos + pruebas E2E + link al deploy
3. **README.md**: Informe breve describiendo:
    * Arquitectura elegida
    * Tecnologías utilizadas
    * Dificultades encontradas y cómo se resolvieron

### Criterios de Evaluación

* Correctitud y completitud de las funcionalidades
* Calidad del código (estructura, buenas prácticas)
* Cobertura y calidad de las pruebas
* Claridad en la documentación
* Creatividad en la interfaz y experiencia de usuario
